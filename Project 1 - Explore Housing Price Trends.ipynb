### As `Neighborhood` column containts 25 unique values with no inherent ordering between them, Lets apply `One Hot Encoding` on that column to encode.


[Detailed Documentation ](https://scikit-learn.org/stable/modules/generated/sklearn.preprocessing.OneHotEncoder.html)

````python
from sklearn.preprocessing import OneHotEncoder


# Initialize the OneHotEncoder
# Setting sparse=False means the output will be a numpy array instead of a sparse matrix
encoder = OneHotEncoder()

# Reshape the 'Type 1' column because fit_transform expects a 2D array
types = df['Neighborhood'].values.reshape(-1, 1)

# Fit the encoder and transform 'Type 1' data to one-hot encoded format
one_hot_encoded_types = encoder.fit_transform(types)

# Convert the sparse matrix to a numpy array
one_hot_encoded_types = one_hot_encoded_types.toarray()

# Create a DataFrame with the encoded data, column names will be the unique categories from 'Type 1'
one_hot_encoded_df = pd.DataFrame(one_hot_encoded_types, columns=encoder.categories_[0])

# Display the first few rows of the one-hot encoded DataFrame
one_hot_encoded_df.head()
````

````python
# Define a function to apply the style
def highlight_ones(val):
    color = 'green' if val == 1 else 'none'
    return f'background-color: {color}'

# Apply the style to the DataFrame
styled_df = one_hot_encoded_df.head().style.applymap(highlight_ones)
styled_df
````

### Applying One-Hot Encoding Using Sklearn

#### 1. **Importing Necessary Libraries:**
- `from sklearn.preprocessing import OneHotEncoder`: This imports the `OneHotEncoder` class from `sklearn.preprocessing`, which is used for one-hot encoding.
- `import pandas as pd`: Imports the Pandas library for data manipulation.

#### 2. **Loading the Dataset:**
- `data = pd.read_csv(file_path)`: Reads the data from the specified CSV file into a Pandas DataFrame.

#### 3. **Initializing the OneHotEncoder:**
- `encoder = OneHotEncoder(sparse=False)`: Creates an instance of the `OneHotEncoder`. The `sparse=False` argument makes the output an array instead of a sparse matrix, which is easier to work with if memory is not an issue.

#### 4. **Preparing Data for Encoding:**
- `types = data['Type 1'].values.reshape(-1, 1)`: Reshapes the 'Type 1' column to be a 2D array, as `fit_transform` requires input to be in this shape.

#### 5. **Fitting and Transforming the Data:**
- `one_hot_encoded_types = encoder.fit_transform(types)`: This line actually performs the one-hot encoding. The encoder learns the categories and transforms the data into a binary (one-hot encoded) format.

#### 6. **Creating a DataFrame from the Encoded Data:**
- `one_hot_encoded_df = pd.DataFrame(one_hot_encoded_types, columns=encoder.categories_[0])`: Constructs a new DataFrame from the encoded data. Column names are set to the unique categories identified by the encoder in the 'Type 1' column.


This approach using `sklearn` is particularly useful when you want to integrate encoding seamlessly with other preprocessing steps in a machine learning pipeline, maintaining consistency and efficiency.

````python
# Select House in two Neighborhood for comparison
Edwards_stats = df.loc[df['Neighborhood'] == 'Edwards', ['OverallQual', 'OverallCond', 'TotRmsAbvGrd', 'GarageCars', 'Fireplaces']].mean().tolist()
OldTown_stats = df.loc[df['Neighborhood'] == 'OldTown', ['OverallQual', 'OverallCond', 'TotRmsAbvGrd', 'GarageCars', 'Fireplaces']].mean().tolist()

# Attributes to compare
categories = ['OverallQual', 'OverallCond', 'TotRmsAbvGrd', 'GarageCars', 'Fireplaces']

# Calculate angles for radar chart
angles = np.linspace(0, 2 * np.pi, len(categories), endpoint=False).tolist()
angles += angles[:1]  # Complete the loop

# Data needs to be repeated at the end to close the radar chart
Edwards_stats += Edwards_stats[:1]
OldTown_stats += OldTown_stats[:1]

# Create radar chart
fig, ax = plt.subplots(figsize=(6, 6), subplot_kw=dict(polar=True))
ax.plot(angles, Edwards_stats, linewidth=2, linestyle='solid', label='Edwards')
ax.fill(angles, Edwards_stats, color='yellow', alpha=0.25)
ax.plot(angles, OldTown_stats, linewidth=2, linestyle='solid', label='OldTown')
ax.fill(angles, OldTown_stats, color='green', alpha=0.25)

# Add labels
ax.set_xticks(angles[:-1])
ax.set_xticklabels(categories)
ax.legend(loc='upper right', bbox_to_anchor=(0.1, 0.1))

plt.title('Attribute Radar Chart for Edwards and OldTown')
plt.show()
````

### 8. Sankey Diagram
- Question it answers: How do Condition and Sales Condition transitions in House differ when considering CentralAir versus Non-CentralAir House, and what are the most common type transitions for each group?

````python
import nbformat
print(nbformat.__version__)
````

````python
import pandas as pd
import plotly.graph_objects as go


# Prepare the data for the Sankey diagram, including CentralAir status correctly
type_pairs = df.groupby(['Condition1', 'SaleCondition']).size().reset_index(name='Count')
CentralAir_status = df.groupby(['Condition1', 'SaleCondition'])['CentralAir'].mean().reset_index()

# Correctly merge the CentralAir status into the type_pairs DataFrame
type_pairs = type_pairs.merge(CentralAir_status, on=['Condition1', 'SaleCondition'], how='left')

# Map types to unique integers for Sankey diagram node indices
type_mapping = {ptype: idx for idx, ptype in enumerate(pd.concat([df['Condition1'], df['SaleCondition'].dropna()]).unique())}

# Define source, target, and value for the Sankey diagram
source = type_pairs['Condition1'].map(type_mapping)
target = type_pairs['SaleCondition'].map(type_mapping)
value = type_pairs['Count']
# Define color based on the average CentralAir status, using a threshold to determine color
color = type_pairs['CentralAir'].apply(lambda x: 'rgba(255, 165, 0, 0.8)' if x > 0.5 else 'rgba(135, 206, 250, 0.8)')

# Generate the Sankey diagram
fig = go.Figure(data=[go.Sankey(
    node=dict(pad=15, thickness=20, line=dict(color='black', width=0.5), label=list(type_mapping.keys())),
    link=dict(source=source, target=target, value=value, color=color)
)])

fig.update_layout(title_text='Sankey Diagram of Condition Type Transitions by CentralAir Status', font_size=10)
# fig.show()
````

Updated Interpretation:

This Sankey diagram will now visually separate the flows of House type transitions based on the proportion of times they involve CentralAir versus Non-CentralAir House. By categorizing the links using different colors:

    Orange: Transitions predominantly involving CentralAir House, indicating rarer or more specialized type combinations.
    Light Blue: Transitions primarily involving Non-CentralAir House, showcasing common or standard type transitions in the House universe.

### 16. Network Graphs

````python
import plotly.graph_objects as go
import networkx as nx
import pandas as pd

# Create the graph
G = nx.Graph()

# Calculate the proportion of CentralAir House for each type
type_CentralAir_proportion = df.groupby('Neighborhood')['CentralAir'].mean().to_dict()

# Add nodes for each House type with the CentralAir proportion as an attribute
all_types = pd.concat([df['Neighborhood'], df['MSZoning'].dropna()]).unique()
for House_type in all_types:
    G.add_node(House_type, CentralAir_proportion=type_CentralAir_proportion.get(House_type, 0))

# Add edges based on dual-type combinations
for _, row in df.dropna(subset=['MSZoning']).iterrows():
    G.add_edge(row['Neighborhood'], row['MSZoning'])

# Node positions using a layout
pos = nx.spring_layout(G)

# Node trace with text for labels
node_trace = go.Scatter(
    x=[pos[node][0] for node in G.nodes()],
    y=[pos[node][1] for node in G.nodes()],
    text=[node for node in G.nodes()],
    mode='markers+text',
    hoverinfo='text',
    marker=dict(
        showscale=True,
        colorscale='YlGnBu',
        color=[G.nodes[node]['CentralAir_proportion'] for node in G.nodes()],
        size=10,
        colorbar=dict(
            thickness=15,
            title='CentralAir Proportion',
            xanchor='left',
            titleside='right'
        ),
    ),
    textposition="top center"
)

# Edge trace
edge_x = []
edge_y = []
for edge in G.edges():
    x0, y0 = pos[edge[0]]
    x1, y1 = pos[edge[1]]
    edge_x.extend([x0, x1, None])
    edge_y.extend([y0, y1, None])

edge_trace = go.Scatter(
    x=edge_x,
    y=edge_y,
    line=dict(width=0.5, color='#888'),
    hoverinfo='none',
    mode='lines'
)

# Create the figure
fig = go.Figure(data=[edge_trace, node_trace],
                layout=go.Layout(
                    title='Network Graph of House Types with CentralAir Proportion',
                    showlegend=False,
                    hovermode='closest',
                    margin=dict(b=0, l=0, r=0, t=40),
                    xaxis=dict(showgrid=False, zeroline=False, showticklabels=False),
                    yaxis=dict(showgrid=False, zeroline=False, showticklabels=False),
                    annotations=[dict(
                        text="Note: Node size and color intensity vary by the proportion of CentralAir House.",
                        showarrow=False,
                        xref="paper", yref="paper",
                        x=0.005, y=-0.002 )]
                ))

fig.show()
````
